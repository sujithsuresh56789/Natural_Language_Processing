## Text Cleaning

### Basic Cleaning-
##### converting to lower case
##### Tokenization
##### Punctuation removal using regex
##### Cleaning stopwords
##### Stemming / Lemmatization

```python
raw_docs=["I am writing some very basic english sentences",
          "I'm just writing it to make people understand the very basics of text preprocessing for NLP",
         "The point is to _learn HOW it _WORKS on #simple #data"]
```

```python
#importing nltk package(natural language toolkit)
import nltk
```

## Converting to lower case

```python
raw_docs=[i.lower() for i in raw_docs]
raw_docs
```

## Tokenization

```python
# Tokenization is the process of splitting sentences into words or group of sentences into sentences
# word tokenize
from nltk.tokenize import word_tokenize
tokenized_docs=[word_tokenize(i) for i in raw_docs]
tokenized_docs
```

```python
# sentence tokenization
from nltk.tokenize import sent_tokenize
tokenized_sentence=[sent_tokenize(i) for i in raw_docs]
tokenized_sentence
```

## Punctuation removal using Regex 

```python
#tokenized_docs = [["This", "is", "a", "sample", "sentence", "."], ["Another", "document", "with", "tokens", "."]]

import re
import string

regex = re.compile('[%s]' % re.escape(string.punctuation))

tokenized_docs_no_punctuation = []

for review in tokenized_docs:
    new_review = []
    for token in review:
        new_token = regex.sub(u'', token)
        if not new_token == u'':
            new_review.append(new_token)
    
    tokenized_docs_no_punctuation.append(new_review)

print(tokenized_docs_no_punctuation)
```

## Cleaning Stopwords

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
tokenized_words_no_stopwords=[]
for doc in tokenized_docs_no_punctuation:
    new=[]
    for word in doc:
        if not word in stopwords.words("english"):
            new.append(word)
    tokenized_words_no_stopwords.append(new)

tokenized_words_no_stopwords
            
```

## Stemming 

```python
from nltk.stem.porter import PorterStemmer
stemmer=PorterStemmer()
stemmed_doc=[]
for doc in tokenized_words_no_stopwords:
    new=[]
    for word in doc:
        new.append(stemmer.stem(word))
    stemmed_doc.append(new)

stemmed_doc
```
